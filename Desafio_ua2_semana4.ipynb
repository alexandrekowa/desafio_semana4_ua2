### 2) Considerando o conceito da galinha dos ovos de ouro, diga quais são os 4 filmes que apresentam maior correlação com todos os outros, a partir da lista de filmes correlacionados a FM selecionado pelo usuário?

```python
matriz_notas_filme = avaliacoes.pivot_table(index='user_id', columns='film_id', values='classification')

# Escolha do filme base (ADICIONA O NOME DO FILME QUE ESTARÁ CORRELACIONANDO)
titulo_base = "Four Rooms (1995)"
item_base = filmes[filmes['title'] == titulo_base]['item_id'].values[0]

# Filmes correlacionados ao FM
correlacoes_filme = matriz_notas_filme.corrwith(matriz_notas_filme[item_base])
correlacoes_filme = pd.DataFrame(correlacoes_filme, columns=['correlacao']).dropna()

# Filtra os 50 mais correlacionados com o filme base
correlacionados_fm = correlacoes_filme.sort_values(by='correlacao', ascending=False).head(50)

# Agora calculamos, para cada um desses 50 filmes, a média da correlação com TODOS os outros filmes
medias_correlacao_filmes = {}
for filme_id in correlacionados_fm.index:
    correlacoes_outros = matriz_notas.corrwith(matriz_notas[filme_id]).dropna()
    media = correlacoes_outros.mean()
    medias_correlacao_filmes[filme_id] = media

# Transforma em DataFrame e ordena
df_medias = pd.DataFrame.from_dict(medias_correlacao_filmes, orient='index', columns=['media_correlacao'])
df_medias = df_medias.sort_values(by='media_correlacao', ascending=False).head(4)

# Junta com os títulos
df_top_filmes = df_medias.merge(filmes, left_index=True, right_on='item_id')

# Resultado final
df_top_filmes[['title', 'media_correlacao']]
```
